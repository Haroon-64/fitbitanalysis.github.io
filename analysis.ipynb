```python
def load_data(path):
    data = []
    for file in os.listdir(path):
        if file.endswith(".csv"):
            print(f"Loading {file}")
            df = pd.read_csv(os.path.join(path, file))
            data.append(df)
    return data
```

```python
# check data issues
print("Daily Activity: ", daily_activity.isnull().sum())
print("Sleep Day: ", sleep_day.isnull().sum())
print("Weight Log Info: ", weight_log_info.isnull().sum())

```

```python
# Number of distinct users in each dataset
activity_users = daily_activity['Id'].nunique()
sleep_users    = sleep_day['Id'].nunique()
weight_users   = weight_log_info['Id'].nunique()

print("Users with activity data:", activity_users)
print("Users with sleep data:", sleep_users)
print("Users with weight data:", weight_users)

# Records per user
activity_days = daily_activity.groupby('Id')['ActivityDate'].nunique()
sleep_days    = sleep_day.groupby('Id')['SleepDay'].nunique()
weight_days   = weight_log_info.groupby('Id')['Date'].nunique()

# Summary table

usage_summary = pd.DataFrame({
    'ActivityDays': activity_days,
    'SleepDays': sleep_days,
    'WeightLogs': weight_days
}).fillna(0)

# Show average and median usage
print("Average days per user:")
print(usage_summary.mean())

print("Median days per user:")
print(usage_summary.median())

# Plot coverage comparison

usage_summary.mean().plot(kind='bar')
plt.ylabel("Average days logged")
plt.title("Average User Engagement by Feature")
plt.show()
```

```python

usage_summary = pd.DataFrame({
    'ActivityDays': daily_activity.groupby('Id')['ActivityDate'].nunique(),
    'SleepDays': sleep_day.groupby('Id')['SleepDay'].nunique(),
    'WeightLogs': weight_log_info.groupby('Id')['Date'].nunique()
}).fillna(0)

# compare user engagement
def classify_user(row):
    features_used = sum([
        row['ActivityDays'] > 0,
        row['SleepDays'] > 0,
        row['WeightLogs'] > 0
    ])
    active_days = row['ActivityDays'] + row['SleepDays'] + row['WeightLogs']

    if features_used == 3 and active_days > 40:
        return "Power User"
    elif features_used >= 2 and active_days > 15:
        return "Engaged User"
    elif features_used >= 1:
        return "Casual User"
    else:
        return "Inactive"

usage_summary['Tier'] = usage_summary.apply(classify_user, axis=1)

# Count users in each tier
tier_counts = usage_summary['Tier'].value_counts()

print(tier_counts)

# Plot distribution
tier_counts.plot(kind='bar')
plt.title("User Engagement Tiers")
plt.ylabel("Number of Users")
plt.show()
```

```python
tier_feature_usage = usage_summary.groupby('Tier')[['ActivityDays', 'SleepDays', 'WeightLogs']].mean()

print(tier_feature_usage)

# Normalize to percentages to see proportions
tier_feature_pct = tier_feature_usage.div(tier_feature_usage.sum(axis=1), axis=0) * 100
print(tier_feature_pct)


plt.figure(figsize=(8,5))
sns.heatmap(tier_feature_pct, annot=True, fmt=".1f", cmap="Blues")
plt.title("Feature Usage Composition by Tier (%)")
plt.ylabel("Tier")
plt.xlabel("Feature")
plt.show()
```
